Initially, I wrote the pipeline for GMB dataset which is in shape of csv file. Later, I decided to use the model on ConLL2003 dataset. Therefore, here I reshape the ConLL2003 data from text to CSV, in the way that fits my model.
Data processing here took about 1 hour. You can download the data from my repository for your ease.

```python
import pandas as pd
import numpy as np
```

```python
def split_text_label(filename):
  f = open(filename)
  split_labeled_text = []
  sentence = []
  for line in f:
    if len(line)==0 or line.startswith('-DOCSTART') or line[0]=="\n":
       if len(sentence) > 0:
         split_labeled_text.append(sentence)
         sentence = []
       continue
    splits = line.split(' ')
    sentence.append([splits[0], splits[1], splits[-1].rstrip("\n")])
  if len(sentence) > 0:
    split_labeled_text.append(sentence)
    sentence = []
  return split_labeled_text
```

```python
split_train = split_text_label('/content/drive/MyDrive/NLP/final/archive/train.txt')
split_test = split_text_label("/content/drive/MyDrive/NLP/final/archive/test.txt")
split_valid = split_text_label("/content/drive/MyDrive/NLP/final/archive/valid.txt")
```

```python
def csv(file):
  import numpy as np
  import pandas as pd
  df = pd.DataFrame(columns={'Sentence #', 'Word','POS', 'Tag'})
  df = df.iloc[:,[0,3,1,2]]
  for i in range(len(file)):
    for j in range(len(file[i])):
        if j == 0:
            df.loc[len(df.index)] = {'Sentence #': 'Sentence: {}'.format(i+1), 'Word':file[i][j][0], 'POS':file[i][j][1] ,'Tag':file[i][j][-1]}
        else:
            df.loc[len(df.index)] = {'Word':file[i][j][0], 'POS':file[i][j][1] ,'Tag':file[i][j][-1]}
  return df
```

```python
test_data = csv(split_test)
test_data
```

```python
valid_data = csv(split_valid)
valid_data
```

```python
train_data = csv(split_train)
train_data
```

```python
data = pd.concat([train_data, valid_data, test_data], axis=0)
data
```

```python
data.to_csv('/content/drive/MyDrive/NLP/final/archive/ConLL2003_data.csv', index=False)
```
